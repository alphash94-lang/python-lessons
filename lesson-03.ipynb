#### `Lists`

###### 1. Accessing List Elements

```python
fruits = ["apples", "banana", "orange"]

# List indexing
print(f'fruits[0]: {fruits[0]}')
print(f'fruits[1]: {fruits[1]}')
print(f'fruits[2]: {fruits[2]}')
print(fruits[1])
```

```python
fruits = ["apple", "banana", "orange"]

# Negative indexing
print(f'fruits[-1]: {fruits[-1]}')
```

```python
fruits = ["apple", "banana", "orange"]

# List and for loop
print("using for loop to iterate over fruits:")
for i in range(len(fruits)):
    print(f'{fruits[i]} is at index {i}')
    print(f'fruits[{i}]: {fruits[i]}')
    print(f'len(fruits[{i}]): {len(fruits[i])}')
```

```python
# changing list items
fruits = ["apple", "banana", "orange"]

fruits[2] = "grape"

print(f'changed fruits: {fruits}')

```

```python
fruits[1] = "melon"

print(fruits)
```

```python
# adding items to a list
fruits = ["apple", "banana"]

fruits.append("orange")

print(f'Adding fruits: {fruits}')
```

```python
# removing items
fruits = ["apple", "banana", "orange"]
fruits.remove("banana")

print(f'removed banana: {fruits}')
```

```python
# popping items
fruits = ["apple", "banana", "orange"]
popped_fruit = fruits.pop(1)
print(f'Popped fruit: {popped_fruit}')
print(f'fruits after popping: {fruits}')
```

###### List Slicing and Useful List Methods

```python
numbers = [10, 20, 30, 40, 50]

print(f'numbers[1:4]: {numbers[1:4]}')
print(f'numbers[:3]: {numbers[:3]}')
print(f'numbers[2:]: {numbers[2:]}')

print(f'numbers[::-1]: {numbers[::-1]}')

```

```python
# sort
numbers = [50, 20, 40, 10, 30]
numbers.sort()
print(f'Sorted numbers: {numbers}')

numbers.sort(reverse=True)
print(f'Sorted numbers in descending order: {numbers}')


```

```python
# reverse
numbers = [50, 20, 40, 10, 30]
numbers.reverse()
print(f'Sorted numbers: {numbers}')
numbers.reverse()
print(f'Reversed numbers again: {numbers}')
print(numbers)
```
