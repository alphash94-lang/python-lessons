#### String Methods

```python
# Uppercase and Lowercases
name = "Alice"
print(f'Upper case: {name.upper()}')  # Output: ALICE
print(f'Lower case: {name.lower()}')  # Output: alice
```

```python
# Capitalize
name_2 = "bob"
print(f'Capitalized: {name_2.capitalize()}')  # Output Bob

# Title Case
full_name = "charlie brown"
print(f'Title Case: {full_name.title()}')  # Output: Charlie Brown
```

#### Some more string methods

```python
# Removing Whitespace
greeting = "   Hello World!   "
print(f'Original: "{greeting}"')
print(f'Stripped: "{greeting.strip()}"')  # Output: "Hello, World!"
```

```python
greeting = "   Hello World!   "
# left strip
print(f'Left Stripped: "{greeting.lstrip()}"')  # Output: "Hello, World!   "
# right strip
print(f'Right Stripped: "{greeting.rstrip()}"')  # Output: "   Hello, World!"
```

```python
# Replacing Substrings
text = "I like cats. Cats are great!"
print(f'Original: "{text}"')
print(f'Replaced: "{text.replace("cats", "dogs")}"')  # Output: "I like dogs. Cats are great!"
```

```python
# Finding Substrings
sentence = "The quick brown fox jumps over the lazy dog."
print(f'Index of "fox": {sentence.find("fox")}')  # Output: 16
# If the substring is not found, find() returns -1
print(f'Index of "cat": {sentence.find("cat")}') # Output: -1
```

```python
# Checking if Text Exists
sentence = "The quick brown fox jumps over the lazy dog."
print(f'Contains "quick": {"quick" in sentence}') # Output: True
print(f'Contains "cat": {"cat" in sentence}')  # Output: False
```

```python
# Splitting and Joining Strings
text = "apple,banana,cherry"
fruits = text.split(",")  # Split the string into a list
print(f'Fruits: {fruits}') # Output: ['apple', 'banana', 'cherry']

text_for_joining = ["Hello", "World"]
joined_text = " ".join(text_for_joining)  # Join the list into a string
print(f'Joined Text: "{joined_text}"')  # Output: "Hello World"
```
