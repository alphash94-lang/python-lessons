#### Functions with Parameters

```python
def greet(name):
    
    print(f"Hello {name}")
    # print("Hello", name)
    # print("Hello" + " " + name)

greet("Alice Kim")
greet("Bob Park")
# greet(name = "Alice Kim")
# greet(name = "Bob Park")
```

#### Multiple Parameters

```python
def introduce(name, age):
    print(f'Hello, my name is {name} and i am {age} years old.')

introduce("Alice", 20)
introduce("Bob", 25)
introduce(age=30, name="Charlie") # you can also use keyword arguements. And order doesn't matter now.
```

#### Function Calling Another Function

```python
def greet():
    print("Hello")

def welcome():
    greet()
    print("welome!")

welcome()
```

```python
def greet():
    print("Hello")

def welcome():
    print("Hello")
    print("welome!")


welcome()
```

#### Returning Multiple Values

```python
def calculate(a, b):

    total = a + b
    difference = a - b
    
    return total, difference

sum_result, diff_result = calculate(10, 5)
# a, b= calculate(10, 5)

print(f"Sum: {sum_result}, Difference: {diff_result}")
```

```python
def calculate(a, b):

    total = a + b
    difference = a - b
    
    return total, difference

sum_result, diff_result = calculate(b = 10, a = 5)

print(f"Sum: {sum_result}, Difference: {diff_result}")
```
